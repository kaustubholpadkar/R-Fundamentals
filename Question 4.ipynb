### Question 4

#### Generate a regular grid between -50 and 50. Construct separate plots of log(x), exp(x), sin(x), sin(2x), sin(x)/cos(x).
#### Examine the cumulative sum of the final function. Experiment with the argument type olf the plot function.

##### 1) log(x)

```r
x = seq(-50, 50, by=0.1)
y = log(x)
```

```r
plot(x, y, "l", xlim = c(-50, 50), ylim = c(0, 20), main = "log(X)")
```

##### 2) exp(x)

```r
x = seq(-50, 50, by=0.1)
y = exp(x)
```

```r
plot(x, y, "l", xlim = c(-50, 50), ylim = c(0, 10000), main = "exp(X)")
```

##### 3) sin(x)

```r
x = seq(-50, 50, by=0.1)
y = sin(x)
```

```r
plot(x, y, "l", xlim = c(-50, 50), ylim = c(-10, 10), main = "sin(X)")
```

##### 4) sin(2x)

```r
x = seq(-50, 50, by=0.1)
y = sin(2 * x)
```

```r
plot(x, y, "l", xlim = c(-50, 50), ylim = c(-10, 10), main = "sin(2X)")
```

##### 5) sin(x) / cos(x)

```r
x = seq(-50, 50, by=0.1)
y = sin(x) / cos(x)
```

```r
plot(x, y, "l", xlim = c(-50, 50), ylim = c(-10, 10), main = "sin(x) / cos(x)")
```
